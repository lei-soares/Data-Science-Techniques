# Regression Models Overview
This notebook summarizes popular regression models.

## Linear Regression
Hyperparameters:
- fit_intercept
- normalize

Strengths: Simple, interpretable.
Weaknesses: Assumes linearity.

## Ridge Regression
Hyperparameters:
- alpha

Strengths: Reduces multicollinearity.
Weaknesses: Still linear.

## Lasso Regression
Hyperparameters:
- alpha

Strengths: Feature selection.
Weaknesses: Can drop useful features.

## ElasticNet
Hyperparameters:
- alpha
- l1_ratio

Strengths: Combines Ridge & Lasso.
Weaknesses: Needs tuning.

## Bayesian Ridge
Hyperparameters:
- alpha_1
- alpha_2

Strengths: Probabilistic.
Weaknesses: Slower.

## Huber Regressor
Hyperparameters:
- epsilon

Strengths: Robust to outliers.
Weaknesses: Sensitive to epsilon.

## RANSAC Regressor
Hyperparameters:
- min_samples

Strengths: Very robust.
Weaknesses: Computational cost.

## Decision Tree Regressor
Hyperparameters:
- max_depth
- min_samples_split

Strengths: Non-linear.
Weaknesses: Overfitting.

## Random Forest Regressor
Hyperparameters:
- n_estimators
- max_depth

Strengths: Strong performance.
Weaknesses: Less interpretable.

## Extra Trees Regressor
Hyperparameters:
- n_estimators

Strengths: Fast, accurate.
Weaknesses: Randomness.

## Gradient Boosting Regressor
Hyperparameters:
- learning_rate
- n_estimators

Strengths: High accuracy.
Weaknesses: Sensitive tuning.

## AdaBoost Regressor
Hyperparameters:
- n_estimators
- learning_rate

Strengths: Simple boosting.
Weaknesses: Noise sensitive.

## XGBoost Regressor
Hyperparameters:
- eta
- max_depth

Strengths: State-of-art.
Weaknesses: Complex.

## LightGBM Regressor
Hyperparameters:
- num_leaves

Strengths: Fast, scalable.
Weaknesses: Overfitting risk.

## CatBoost Regressor
Hyperparameters:
- depth

Strengths: Handles categorical.
Weaknesses: Slower training.

## SVR
Hyperparameters:
- C
- kernel

Strengths: Flexible.
Weaknesses: Poor scaling.

## KNN Regressor
Hyperparameters:
- n_neighbors

Strengths: Simple.
Weaknesses: Memory heavy.

## Gaussian Process Regressor
Hyperparameters:
- kernel

Strengths: Uncertainty estimates.
Weaknesses: Very slow.

## Poisson Regressor
Hyperparameters:
- alpha

Strengths: Count data.
Weaknesses: Limited use.

## Tweedie Regressor
Hyperparameters:
- power

Strengths: Flexible distributions.
Weaknesses: Hard tuning.
