# Embedding-Based Topic Models

    Semantic Topic Discovery Beyond Bag-of-Words
    
## Objective

This notebook provides a structured and practical introduction to embedding-based topic modeling, covering:

- Why classical topic models break down

- Sentence and document embeddings

- Clustering in embedding space

- Topic representation from embeddings

- Strengths, risks, and governance considerations

It answers:

    How can we discover semantic topics when word-count assumptions are too restrictive?

## Why Embedding-Based Topic Models?

LDA assumes:

- Bag-of-words

- Exchangeability

- Sparse word counts

These assumptions break when:

- Texts are short (complaints, tickets)

- Synonyms dominate

- Context matters

- Vocabulary is noisy

- Embeddings solve semantic similarity, not probability modeling.

## Business Context – Finance & Banking

Embedding-based topic models are used for:

Short customer complaints

Chat and call transcripts

Social media monitoring

AML alert narratives

Compliance text triage

     They trade interpretability and statistical grounding for semantic power.

## Dataset



```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns


df = pd.read_csv("D:/GitHub/Data-Science-Techniques/datasets/Unsupervised/bank_complaints.csv")

df.head()
```

# Minimal Text Preprocessing

Embedding models do NOT require heavy cleaning.

## Cleaning

```python
import re

def light_clean(text):
    text = text.lower()
    text = re.sub(r"\s+", " ", text)
    return text.strip()

df["clean_text"] = df["text"].apply(light_clean)
```

✔ No stemming
✔ No stopword removal
✔ Context preserved

## Text Embeddings

We use Sentence Transformers for document-level embeddings.

Common choices:

- all-MiniLM-L6-v2 (fast, strong baseline)

- mpnet-base-v2 (higher quality)

### Generate Embeddings

```python
# !pip install sentence-transformers
```

```python
from sentence_transformers import SentenceTransformer

model = SentenceTransformer("all-MiniLM-L6-v2")

embeddings = model.encode(
    df["clean_text"].tolist(),
    show_progress_bar=True
)
```

```python
#(n_documents, embedding_dimension)
```

## Why Clustering Embeddings?

Embeddings place documents in **semantic space.**

Topics = **dense regions in embedding space.**

Clustering options:

- K-Means (simple)

- Hierarchical (interpretable)

- HDBSCAN (noise-aware, recommended)

## HDBSCAN for Topic Discovery

HDBSCAN is preferred because:

- No fixed K

- Handles noise

- Variable topic sizes

- Works well with embeddings

### HDBSCAN Clustering

```python
# !pip install hdbscan
```

```python
import hdbscan

clusterer = hdbscan.HDBSCAN(
    min_cluster_size=30,
    metric="euclidean",
    cluster_selection_method="eom"
)

df["topic"] = clusterer.fit_predict(embeddings)
```

## Topic Distribution
df["topic"].value_counts()


- -1 → noise / miscellaneous

- 0, 1, 2… → discovered topics

✔ Noise is expected and healthy

## Topic Interpretation Strategy

Embedding clusters do not produce words directly.

Interpretation requires:

- Representative documents

- Keyword extraction per cluster

- Human review

## Extract Representative Texts
### Sample Documents per Topic

```python
for topic in sorted(df["topic"].unique()):
    if topic == -1:
        continue
    print(f"\nTopic {topic}")
    sample_texts = df[df["topic"] == topic]["text"].head(3)
    for t in sample_texts:
        print("-", t[:120])
```

## Keyword Extraction (TF-IDF per Cluster)
### TF-IDF Topics

```python
from sklearn.feature_extraction.text import TfidfVectorizer

def extract_keywords(texts, top_n=10):
    tfidf = TfidfVectorizer(stop_words="english", max_features=5000)
    X = tfidf.fit_transform(texts)
    scores = X.mean(axis=0).A1
    terms = tfidf.get_feature_names_out()
    return sorted(
        zip(terms, scores),
        key=lambda x: x[1],
        reverse=True
    )[:top_n]

for topic in sorted(df["topic"].unique()):
    if topic == -1:
        continue
    keywords = extract_keywords(df[df["topic"] == topic]["clean_text"])
    print(f"\nTopic {topic} keywords:")
    print([k for k, _ in keywords])
```

# Visualization (UMAP)
## UMAP Projection

```python
import umap

umap_model = umap.UMAP(
    n_neighbors=15,
    min_dist=0.1,
    random_state=42
)

X_umap = umap_model.fit_transform(embeddings)
```

### Plot Topics

```python
plt.figure(figsize=(6, 6))
sns.scatterplot(
    x=X_umap[:, 0],
    y=X_umap[:, 1],
    hue=df["topic"],
    palette="tab10",
    alpha=0.7
)
plt.title("Embedding-Based Topics (UMAP)")
plt.show()
```

Visualization ≠ validation

## Comparison: LDA vs Embedding Topics

| Aspect           | LDA           | Embedding-Based |
| ---------------- | ------------- | --------------- |
| Assumption       | Probabilistic | Semantic        |
| Interpretability | High          | Medium          |
| Short texts      | Poor          | Strong          |
| Noise handling   | Weak          | Strong          |
| Governance       | Easier        | Harder          |


## Governance & Risk Considerations (Critical)

Embedding-based topics:
- `[Con] -` Are not probabilistic
- `[Con] -` Are model-dependent
- `[Con] -` Are harder to audit
- `[Con] -` Can drift silently

✔ Always document:

- Model version

- Training data

- Update cadence

- Human review process

## Common Mistakes (Avoided)

- `[Cons] -` Treating clusters as labels
- `[Cons] -` Skipping human validation
- `[Cons] -` Ignoring noise points
- `[Cons] -` Over-trusting visual separation
- `[Cons] -` Using embeddings blindly in production

## Summary Table







| Component      | Recommendation              |
| -------------- | --------------------------- |
| Embeddings     | Sentence-level              |
| Clustering     | HDBSCAN                     |
| Interpretation | Hybrid (keywords + samples) |
| Use case       | Short, semantic text        |
| Governance     | Mandatory                   |


 ## Key Takeaways

- Embedding-based topics capture semantics

- They are discovery tools, not classifiers

- Interpretation is external to the model

- Noise is expected and valuable

- Governance is essential in finance
