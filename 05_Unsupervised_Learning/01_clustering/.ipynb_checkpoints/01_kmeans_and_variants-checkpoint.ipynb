# K-Means and Variants
    Customer Segmentation with Distance-Based Clustering
## Objective

This notebook provides a systematic and practical treatment of K-Means clustering, covering:

- When clustering is useful in business

- K-Means assumptions and limitations

- Choosing the number of clusters

- K-Means variants and improvements

- Interpreting clusters for decision-making

It answers:

    How can we segment customers meaningfully when no labels exist?

## Business Context â€“ Why Clustering in Finance?

In banking and finance, clustering is used for:

- Customer segmentation

- Risk profiling

- Marketing personalization

- Portfolio analysis

- Strategy design (without targets)

Clustering is exploratory, not predictive.

## Imports and Dataset

```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns


df = pd.read_csv("D:/GitHub/Data-Science-Techniques/datasets/Unsupervised/synthetic_bank_customer_segmentation.csv")

df.head()
```

## Dataset Description



| Feature            | Meaning             |
| ------------------ | ------------------- |
| age                | Customer age        |
| annual_income      | Annual income       |
| avg_monthly_spend  | Average spend       |
| credit_utilization | Credit usage ratio  |
| tenure_years       | Relationship length |
| num_products       | Products held       |
| risk_score         | Internal score      |


# Select Clustering Features

Clustering requires numerical, comparable features.

## Feature Selection

```python
features = [
    "age",
    "annual_income",
    "avg_monthly_spend",
    "credit_utilization",
    "tenure_years",
    "num_products",
    "risk_score"
]

X = df[features]
```

# Why Scaling Is Mandatory

K-Means is distance-based.

Without scaling:

- Income dominates distance

- Other features become irrelevant
## Scaling

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

# MODEL

## Baseline K-Means
Assumptions

- Spherical clusters

- Equal variance

- Similar density

## Fit K-Means

```python
from sklearn.cluster import KMeans

kmeans = KMeans(
    n_clusters=4,
    random_state=42,
    n_init=20
)

clusters = kmeans.fit_predict(X_scaled)
df["cluster"] = clusters
```

### Cluster Size Inspection

```python
df["cluster"].value_counts().sort_index()
```

 Extremely imbalanced clusters often signal:

- Wrong K

- Non-spherical structure

### Choosing K (Elbow Method)

```python
inertia = []

for k in range(2, 11):
    km = KMeans(n_clusters=k, random_state=42, n_init=20)
    km.fit(X_scaled)
    inertia.append(km.inertia_)

plt.plot(range(2, 11), inertia, marker="o")
plt.xlabel("Number of Clusters (k)")
plt.ylabel("Inertia")
plt.title("Elbow Method")
plt.show()
```

- Look for diminishing returns
- Elbow not always clear

### Silhouette Analysis

```python
from sklearn.metrics import silhouette_score

sil_scores = {}

for k in range(2, 11):
    km = KMeans(n_clusters=k, random_state=42, n_init=20)
    labels = km.fit_predict(X_scaled)
    sil_scores[k] = silhouette_score(X_scaled, labels)

pd.Series(sil_scores)
```

-  Measures separation quality
-  Higher is better

### Final K Selection

Assume:

```python
k = 4
```

## Cluster Profiling (Critical Step)

```python
cluster_profile = (
    df
    .groupby("cluster")[features]
    .mean()
    .round(2)
)

cluster_profile
```

## Business Interpretation Example

| Cluster | Description                         |
| ------- | ----------------------------------- |
| 0       | Older, conservative, low usage      |
| 1       | High income, low risk, loyal        |
| 2       | Young, low income, high utilization |
| 3       | Medium income, active spenders      |



## Visualization (2D Projection)

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)

plt.figure(figsize=(6, 6))
sns.scatterplot(
    x=X_pca[:, 0],
    y=X_pca[:, 1],
    hue=df["cluster"],
    palette="tab10"
)
plt.title("Clusters (PCA Projection)")
plt.show()
```

PCA is for visualization only.

## K-Means++ Initialization

Default in sklearn, but conceptually important.

- Faster convergence
- Better stability

### Mini-Batch K-Means (Large Data)

```python
from sklearn.cluster import MiniBatchKMeans

mbk = MiniBatchKMeans(
    n_clusters=4,
    batch_size=1000,
    random_state=42
)

df["cluster_mini"] = mbk.fit_predict(X_scaled)
```

- `[pros] -` Scales to millions of rows
- `[cons] -` Slightly less precise

## Step 12 â€“ K-Means Limitations (Critical)

- `[cons] -` Sensitive to outliers
- `[cons] -` Requires K upfront
- `[cons] -` Assumes spherical clusters
- `[cons] -` Fails with varying density

> Not always the right clustering algorithm.

## Common Mistakes (Avoided)

- `[cons] -` Skipping scaling
- `[cons] -` Using categorical variables directly
- `[cons] -` Interpreting clusters as truth
- `[cons] -` Ignoring business meaning
- `[cons] -` Overfitting segmentation

## Summary Table


| Aspect              | K-Means |
| ------------------- | ------- |
| Scalability         | High    |
| Interpretability    | Medium  |
| Outlier Sensitivity | High    |
| Business Use        | High    |
| Assumptions         | Strong  |


## Key Takeaways

- K-Means is simple but powerful

- Scaling is mandatory

- K selection requires diagnostics

- Profiling > labels

- Business meaning is the goal
