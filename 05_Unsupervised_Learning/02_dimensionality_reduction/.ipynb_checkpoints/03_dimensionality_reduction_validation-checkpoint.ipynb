# Dimensionality Reduction Validation

    Ensuring Compression Preserves Signal, Stability, and Business Meaning

##  Objective

This notebook provides a systematic framework for validating dimensionality reduction techniques, covering:

- Variance and information retention

- Downstream task performance

- Stability under perturbations

- Sensitivity to preprocessing

- Interpretability and business alignment

It answers:

    How do we know dimensionality reduction is helping—and not hiding signal?

##  Why Validation Is Necessary

Dimensionality reduction can:

- Remove noise ✔

- Remove signal ❌

- Distort decision boundaries ❌

Validation ensures:
- ✔ Compression improves or preserves outcomes
- ✔ Interpretability is not lost
- ✔ Results are stable and reproducible

##  Imports and Dataset

```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns


df = pd.read_csv("D:/GitHub/Data-Science-Techniques/datasets/Unsupervised/synthetic_bank_customer_segmentation.csv")

df.head()
```

```python
features = [
    "age",
    "annual_income",
    "avg_monthly_spend",
    "credit_utilization",
    "tenure_years",
    "num_products",
    "risk_score"
]

X = df[features]

# Standardization
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

# Variance Retention Validation (PCA)
### PCA Fit

```python
from sklearn.decomposition import PCA

pca = PCA()
X_pca = pca.fit_transform(X_scaled)
```

## Reconstruction Error (PCA)

Reconstruction error quantifies information loss.

### Reconstruction Error

```python
from sklearn.metrics import mean_squared_error

pca_90 = PCA(n_components=0.9)
X_reduced = pca_90.fit_transform(X_scaled)
X_reconstructed = pca_90.inverse_transform(X_reduced)

reconstruction_error = mean_squared_error(X_scaled, X_reconstructed)
reconstruction_error
```

✔ Lower is better

✔ Compare across component counts

## Downstream Task Validation (Clustering)
### K-Means on Original vs Reduced Space

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score

# Original
labels_orig = KMeans(n_clusters=4, random_state=42).fit_predict(X_scaled)
sil_orig = silhouette_score(X_scaled, labels_orig)

# PCA-reduced
labels_pca = KMeans(n_clusters=4, random_state=42).fit_predict(X_reduced)
sil_pca = silhouette_score(X_reduced, labels_pca)

sil_orig, sil_pca
```

✔ Reduction should preserve or improve separation

## Stability Under Resampling
### Stability Test (PCA Components)

```python
from sklearn.metrics import adjusted_rand_score

rng = np.random.RandomState(42)
stability_scores = []

base_labels = labels_pca

for i in range(20):
    sample_idx = rng.choice(len(X_scaled), size=int(0.8 * len(X_scaled)), replace=False)
    X_sample = X_scaled[sample_idx]

    X_sample_pca = pca_90.fit_transform(X_sample)
    labels_sample = KMeans(n_clusters=4, random_state=i).fit_predict(X_sample_pca)

    stability_scores.append(
        adjusted_rand_score(
            base_labels[sample_idx],
            labels_sample
        )
    )

np.mean(stability_scores), np.std(stability_scores)
```

✔ High mean

✔ Low variance

## Sensitivity to Scaling
### Robust Scaling Comparison

```python
from sklearn.preprocessing import RobustScaler

X_robust = RobustScaler().fit_transform(X)
X_robust_pca = PCA(n_components=0.9).fit_transform(X_robust)

labels_robust = KMeans(n_clusters=4, random_state=42).fit_predict(X_robust_pca)
silhouette_score(X_robust_pca, labels_robust)
```

✔ PCA results must be scaling-aware

## Interpretability Validation
###  Loadings

```python
loadings = pd.DataFrame(
    pca.components_.T,
    columns=[f"PC{i+1}" for i in range(len(features))],
    index=features
)

loadings
```

# Nonlinear Method Validation (t-SNE / UMAP)

Nonlinear methods cannot be validated via:
- `[No] -` Reconstruction error
- `[No] -` Variance

Instead validate:

- Visual stability

- Consistency across seeds

- Alignment with known segments

 ## Example — UMAP Stability Check

```python
import umap

embeddings = []

for seed in [0, 10, 42]:
    umap_model = umap.UMAP(random_state=seed)
    embeddings.append(umap_model.fit_transform(X_scaled))
```

## Business Validation Checklist


| Question                   | Yes/No |
| -------------------------- | ------ |
| Signal preserved?          | ✔      |
| Noise reduced?             | ✔      |
| Interpretability retained? | ✔      |
| Downstream task improved?  | ✔      |
| Stable over time?          | ✔      |


## Common Mistakes (Avoided)

- `[No] -` Blindly trusting variance thresholds
- `[No] -` Using nonlinear embeddings as features
- `[No] -` Skipping downstream validation
- `[No] -` Ignoring stability
- `[No] -` Over-compressing features

## Summary Table

| Validation Aspect | Method               |
| ----------------- | -------------------- |
| Information loss  | Reconstruction error |
| Separation        | Silhouette           |
| Stability         | Resampling           |
| Sensitivity       | Scaling tests        |
| Interpretability  | Loadings             |


## Key Takeaways

- Dimensionality reduction must be validated

- PCA validation is quantitative

- Nonlinear methods require qualitative checks

- Stability is as important as performance

- Business meaning is non-negotiable
