# Feature Stability and Drift Checks
     Ensuring Feature Reliability Over Time

## Objective

This notebook provides a systematic treatment of **feature stability and drift detection**, covering:

- Why feature stability matters
- Data drift vs concept drift
- Population Stability Index (PSI)
- Distributional comparisons over time
- Stability-aware feature selection

It answers:

**How do we detect when engineered features stop behaving as expected?**


## Why Feature Stability Matters

Unstable features cause:

- Model degradation
- Silent performance decay
- Invalid interpretations
- Increased operational risk

Feature engineering does not end at training time.


## Imports and Dataset



```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns


df = pd.read_csv("../datasets/03_feature_engineering/synthetic_customer_churn_temporal.csv",
                parse_dates=["event_date"]
)


df.sort_values(["customer_id", "event_date"], inplace=True)
df.head()
```

## Step 1 – Define Time Windows

We compare **reference** vs **current** periods.


```python
reference_period = df[df["event_date"] < "2023-01-01"]
current_period = df[df["event_date"] >= "2023-01-01"]
```

## Step 2 – Numeric Feature Stability (Visual)


```python
feature = "avg_monthly_usage"

plt.figure(figsize=(8, 4))
sns.kdeplot(reference_period[feature], label="Reference", fill=True)
sns.kdeplot(current_period[feature], label="Current", fill=True)
plt.title(f"Distribution Drift: {feature}")
plt.legend()
plt.show()
```

## Step 3 – Population Stability Index (PSI)

PSI quantifies distribution shift.


```python
def calculate_psi(expected, actual, bins=10):
    breakpoints = np.linspace(0, 100, bins + 1)
    expected_perc = np.percentile(expected, breakpoints)
    actual_perc = np.percentile(actual, breakpoints)

    psi = 0
    for i in range(len(expected_perc) - 1):
        e = ((expected >= expected_perc[i]) & (expected < expected_perc[i + 1])).mean()
        a = ((actual >= actual_perc[i]) & (actual < actual_perc[i + 1])).mean()
        if e > 0 and a > 0:
            psi += (e - a) * np.log(e / a)
    return psi
```
